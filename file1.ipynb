```python
# Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn
sns.set_style("whitegrid")

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a bar plot for the population of each state
sns.barplot(x='State name', y='Population', data=data, estimator=sum, ci=None)  

plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.xlabel('State Name')
plt.ylabel('Population')
plt.title('Population of Each State in India (2011 Census)')
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
# Create a bar plot for the population of each district
plt.figure(figsize=(12, 6))
sns.barplot(x='District name', y='Population', data=data, estimator=sum, ci=None)
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.xlabel('District Name')
plt.ylabel('Population')
plt.title('Population of Each District in India (2011 Census)')     
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
age_group_data = data.groupby('State name')[['Age_Group_0_29','Age_Group_30_49','Age_Group_50']].sum().sort_values(by='Age_Group_0_29', ascending=False)
print(age_group_data)

age_group_data.plot(kind='bar', figsize=(12, 6))
plt.title('Population Distribution by Age Group in India (2011 Census)')   
plt.xlabel('State Name')
plt.ylabel('Population')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.legend(title='Age Group')
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
# Create a pie chart for the population distribution by age group
plt.figure(figsize=(8, 8))
plt.pie(age_group_data.sum(), labels=age_group_data.columns, autopct='%1.1f%%', startangle=140)
plt.title('Population Distribution by Age Group in India (2011 Census)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
# Create a heatmap for the population distribution by age group across states
plt.figure(figsize=(12, 8))
sns.heatmap(age_group_data, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Population Distribution by Age Group in India (2011 Census)')
plt.xlabel('State Name')
plt.ylabel('Age Group')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()

```

```python
# Import necessary libraries for visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn
sns.set_style("whitegrid")
```

```python


# 1. Population distribution by state
plt.figure(figsize=(12, 6))
sns.barplot(x=sorted_pop_state_wise.index, y=sorted_pop_state_wise.values, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel('State Name')
plt.ylabel('Population')
plt.title('Population Distribution by State (2011 Census)')
plt.tight_layout()
plt.show()
```

```python


# 2. Age group distribution across states
plt.figure(figsize=(12, 6))
age_group_data.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='coolwarm')
plt.title('Age Group Distribution Across States')
plt.xlabel('State Name')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.legend(title='Age Group')
plt.tight_layout()
plt.show()

# # 3. Correlation heatmap for numerical columns in the dataset
# plt.figure(figsize=(12, 8))
# correlation_matrix = data.corr()
# sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', cbar=True)
# plt.title('Correlation Heatmap of Numerical Features')
# plt.tight_layout()
# plt.show()

# 4. Distribution of population by district
plt.figure(figsize=(12, 6))
sns.histplot(data['Population'], kde=True, bins=30, color='blue')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Distribution of Population by District')
plt.tight_layout()
plt.show()

# 5. Boxplot for population distribution by state
plt.figure(figsize=(12, 6))
sns.boxplot(x='State_Name', y='Population', data=data, palette='Set3')
plt.xticks(rotation=90)
plt.xlabel('State Name')
plt.ylabel('Population')
plt.title('Boxplot of Population Distribution by State')
plt.tight_layout()
plt.show()

# 6. Pairplot for selected numerical columns
selected_columns = ['Population', 'Male', 'Female', 'Literate', 'Workers']
sns.pairplot(data[selected_columns], diag_kind='kde', corner=True)
plt.suptitle('Pairplot of Selected Numerical Features', y=1.02)
plt.show()

# 7. Pie chart for age group distribution
age_group_totals = age_group_data.sum()
plt.figure(figsize=(8, 8))
plt.pie(age_group_totals, labels=age_group_totals.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Age Group Distribution (2011 Census)')
plt.axis('equal')
plt.show()
```

```python
# 1. Data Shape + Column Overview
print("Data Shape:", data.shape)
print("Columns Overview:\n", data.columns)
print("\nData Types:\n", data.dtypes)
print("\nData Info:\n")
print(data.info())

# 2. Missing Values Bar Plot
plt.figure(figsize=(10, 6))
data.isnull().sum().plot(kind='bar', title='Missing Values per Column', color='skyblue')
plt.xlabel('Columns')
plt.ylabel('Count of Missing Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Data Type Distribution (Optional Pie Chart)
data_types = data.dtypes.value_counts()
plt.figure(figsize=(8, 8))
plt.pie(data_types, labels=data_types.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Data Type Distribution')
plt.axis('equal')
plt.show()

# 4. Basic Histogram of Selected Columns
selected_columns = ['Population', 'Male', 'Female', 'Literate']
data[selected_columns].hist(bins=30, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Selected Numeric Columns')
plt.tight_layout()
plt.show()
```

```python
# 5. Raw Value Counts (Top 10 Most Common States)
plt.figure(figsize=(10, 6))
data['S'].value_counts().head(10).plot(kind='bar', color='orange', title='Top 10 Most Common States')
plt.xlabel('State Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. Categorical Distributions (Pie Chart of Entries by State)
state_counts = data['State_name'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Entries by State')
plt.axis('equal')
plt.show()

# 7. Dirty Data Spotting (Unique Values in Selected Columns)
print("\nUnique Values in 'State name':\n", data['State_name'].unique())
print("\nUnique Values in 'District name':\n", data['District_name'].unique())
```
